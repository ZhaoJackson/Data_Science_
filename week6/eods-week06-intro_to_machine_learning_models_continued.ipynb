```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from mlxtend.plotting import plot_decision_regions

from sklearn.linear_model import LinearRegression,LogisticRegression

sns.set_style('darkgrid')
%matplotlib inline
```

```python
def my_plot_decision_regions(X,y,model,figsize=(5,5),ax=None):
    '''Plot classifier decision regions, classification predictions and training data'''
    if not ax:
        fig,ax = plt.subplots(1,1,figsize=figsize)
    # use mlxtend plot_decision_regions
    model = model.fit(X.values,y.values)
    plot_decision_regions(X.values,y.values,model,ax=ax)
    ax.set_xlabel(X.columns[0]); ax.set_ylabel(X.columns[1]);
    
def my_plot_regression(X,y,model,label='yhat',figsize=(5,5),ax=None):
    '''Plot regression predictions and training data'''
    # generate test data and make predictions
    X_test = np.linspace(X.iloc[:,0].min(),X.iloc[:,0].max(),1000).reshape(-1,1)
    model = model.fit(X.values,y.values)
    y_hat = model.predict(X_test)
    fig,ax = plt.subplots(1,1,figsize=figsize)
    ax.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
    ax.plot(X_test, y_hat, color="cornflowerblue", label=label, linewidth=2)
    ax.set_xlabel(X.columns[0]); ax.set_ylabel(y.name); ax.legend();
```

# Wine as Binary Classification

```python
df_wine = pd.read_csv('../data/wine_dataset.csv',usecols=['alcohol','ash','proline','hue','class'])
# rename 'class' as 'target', since class is a reserved python word
df_wine = df_wine.rename({'class':'target'},axis=1)

df_wine_2c = df_wine[df_wine.target < 2] # only keep classes 0 and 1

X_2c = df_wine_2c[['proline','hue']]
y_2c = df_wine_2c['target']

zscore = lambda x: (x-x.mean()) / x.std()

X_2c_zscore = X_2c.apply(zscore,axis=0)
alcohol_2c_zscore = zscore(df_wine_2c.alcohol)

y_2c.value_counts().sort_index()
```

- by df_wine.target < 2, we convert the multi-class classification problem into a binary classification problem.
- IV df_wine_2c[['proline','hue']] will be 2-D
- DV df_wine_2c['target'] will be 1-D
- In lambda x: (x-x.mean()) / x.std(): 
    - x.mean() is the mean of the data where lambda x is to extract column variables

# KNN in sklearn

- KNN is a baseline model without hyperparameter tuning.
- KNN is a simple algorithm that works well for classification problems with small datasets.
- KNN is a non-parametric algorithm that does not make any assumptions about the underlying data distribution.

### Steps of KNN

- The algorithm uses the training data directly to make predictions.
- Given the training data, we predict the label of new data points by projecting them into the same space as the training data.
- It estimates the distance between this new data point and all other points in the training set such as Euclidian distance.
- The k-nearest neighbors are identified based on the calculated distances. The parameter ‘k’ determines how many neighbors influence the prediction.
- For classification, it uses majority voting among the nearest neighbors to decide the label for the new data point.
- In regression, it typically averages the values of the nearest neighbors.


```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_2c,y_2c)

my_plot_decision_regions(X_2c,y_2c,knn)
```

# Curse of Dimensionality

The more dimensions, the less likely points are "close" to each other.

```python
# From Data Science From Scratch by Joel Grus

def random_distances(dim,num_pairs=10_000):
    return np.sqrt(np.square(np.random.rand(num_pairs,dim) - np.random.rand(num_pairs,dim)).sum(axis=1))

# calculate average and minimum distance for 1 to 100 dimensions
dimensions = range(1,100)
avg_distances = []
min_distances = []
min_avg_ratio = []
np.random.seed(0)
for d in dimensions:
    distances = random_distances(d)
    avg_distances.append(distances.mean())
    min_distances.append(distances.min())
    min_avg_ratio.append(distances.min() / distances.mean())
```

- more neigbors will have more noise since they are not as well-separated from the rest of the data and each neighbor will have a more different set of features.
- since KNN can only operate in 2-D space, it will be a shortcome when dealing with high-dimensional data where data will elicit more precise and complex relationships.

# Curse of Dimensionality Cont.


```python
fig,ax = plt.subplots(1,2,figsize=(16,7))
ax[0].plot(dimensions,avg_distances,label='avg distance');
ax[0].plot(dimensions,min_distances,label='min distance');
ax[0].legend()
ax[0].set_title('average and min distance'); ax[0].set_xlabel('dimensions'); ax[0].set_ylabel('euclidean distance');
ax[1].plot(dimensions,min_avg_ratio)
ax[1].set_title('min / average distance ratio'); ax[1].set_xlabel('dimensions'); ax[1].set_ylabel('euclidean distance');
```

### Purpose and Greedy Algorithm for Decision Tree Classifier
Purpose of decision tree classifier
 - it tends to maximize information gain, which is a measure of the reduction in entropy after splitting the data based on a particular attribute.
 - it also tends to reduce impurity, which is a measure of the homogeneity of the target variable after splitting the data.

Greedy Algorithm for Decision Tree Classifier
 1. Start with the root node and select the attribute that results in the highest information gain.
 2. Split the data based on the selected attribute.
 3. Recursively apply the above two steps to each of the resulting child nodes until the stopping criterion is met.
 - The stopping criterion for decision tree classifier is when the information gain is zero or when the number of samples in a node is less than a certain threshold.
 
reason for greedy
 - it is easy to make tree overfitting since the algo tends to choose the attribute that results in the highest information gain at each step.

# Stacking
<br>

<div align="center"><img src="images/modelstacking.png" width="900px"></div>

<font size=5>From https://blogs.sas.com/content/subconsciousmusings/2017/05/18/stacked-ensemble-models-win-data-science-competitions/</font>

- stacking is a technique for combining multiple models to improve predictive performance.
- it involves training a meta-model, which uses the predictions of the base models as input features.
- the meta-model is trained to predict the target variable using the predictions of the base models as input features.

### model stacking
- it aims to aggregate the models together
- but we must be careful all the models should be focus on either classification or regression

# Stacking for Classification

```python
from sklearn.ensemble import StackingClassifier

ensemble = [('lr',LogisticRegression(max_iter=1000)),
            ('dt',DecisionTreeClassifier(max_depth=3)),
            ('knn',KNeighborsClassifier(n_neighbors=3))]

stackc = StackingClassifier(estimators=ensemble,
                         final_estimator=LogisticRegression())
stackc.fit(X_2c_zscore,y_2c)

my_plot_decision_regions(X_2c_zscore,y_2c,stackc,figsize=(6,6))
```

# Stacking for Regression

```python
from sklearn.ensemble import StackingRegressor

ensemble = [('lr',LinearRegression()),
            ('dt',DecisionTreeRegressor(max_depth=3)),
            ('knn',KNeighborsRegressor(n_neighbors=6))]

stackr = StackingRegressor(estimators=ensemble,
                           final_estimator=LinearRegression())
stackr.fit(X_2c_zscore[['proline']],alcohol_2c_zscore)

my_plot_regression(X_2c_zscore[['proline']],alcohol_2c_zscore,stackr,figsize=(6,6))
```

# One Vs. Rest (OvR) Classification For Multiclass

- What about other models (eg Perceptron)?
<br>

- Can use any binary classifier for Multiclass classification by training multiple models:

 - model 1 : class 1 vs (class 2 and class 3)
 - model 2 : class 2 vs (class 1 and class 3)
 - model 3 : class 3 vs (class 1 and class 2)
<br>
<br>

- Then
 - Predict $\hat{y}$ using the model with highest $P(y=\hat{y}\mid x)$, or distance from boundary, or ...
<br>
<br>
